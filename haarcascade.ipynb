```
import numpy as np
import cv2 as cv 
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```
urls = ["rostos.jpg"]

for url in urls:
  # carregando a imagem a partir da url
  image = io.imread(url) 
  # transformando de bgr para rgba
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  
  # concatenando as duas imagens
  final_frame = cv.hconcat((image, image_2)) 

  #mostrando as imagens concatenadas
  cv2_imshow(final_frame)

  # transformando para escala de cinza
  image_2_PB = cv.cvtColor( image_2, cv.COLOR_BGR2GRAY)

  # mostrando a imagem em escala de cinza
  cv2_imshow(image_2_PB)


```

```
# carregando o classificador
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')


# aqui vai executar a dettecao de imagens
faces = face_cascade.detectMultiScale(image_2_PB, scaleFactor = 1.05, minNeighbors = 7, minSize = (30,30), flags = cv.CASCADE_SCALE_IMAGE)

# Desenha retangulos na iamgem original (colorida) para cada face encontrada:
for (x,y,w,h) in faces:
      # desenhando retangulo
     cv.rectangle(image_2,(x,y),(x+w,y+h),(255,0,0),2)
     # definindo radius e fazendo casting para int
     radius = int(w / 2)
     # centro x e fazendo casting para int
     centerX = int((x + w) / 2)
     # centro y e fazendo casting para int
     centerY = int((y + h)/ 2)
     center_coordinates = (centerX,centerY)
     # cor vermelha
     color = (0,0,255)
     #desenhando circulo
     cv.circle(image_2, center_coordinates, radius, color, 2)


# mostrando a imagem
cv2_imshow(image_2)
```
